```python
'如何使用已經訓練好的模型？ Ans1:'

from yolo import YOLO
log_dir='logs/000/'
classes_path = 'model_data/voc_classes.txt'
yolo_model = YOLO(model_path=log_dir + 'trained_weights_final.h5', classes_path=classes_path)
r_image = yolo_model.detect_image(image)

r_image
```
